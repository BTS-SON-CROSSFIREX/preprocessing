```python
def preprocess(str):
    # 불필요한 공백 삭제
    str = str.strip(" ")
    # 대문자 소문자로 바꿈
    str = str.lower()
    # @이나 RT로 시작하는 부분 다 지움
    while(1):
        if(str[0:1] == '@' or str[0:2] == 'RT') :
            str = ' '.join(str.split()[1:])
        else : break
    # #, @, https 지움
    str = re.sub('#|@|http.*|\(|\)|\"|\*', '', str)
    # 공백 문자 실제 공백으로 대체
    str = re.sub('\s|-|:|;|&|/', ' ', str)
    # 물음표, 느낌표 온점으로 대체
    str = re.sub('!|\?', '.', str)
    
    # 이모지 제거
    emoj = re.compile("["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
        u"\U00002500-\U00002BEF"  # chinese char
        u"\U00002702-\U000027B0"
        u"\U00002702-\U000027B0"
        u"\U000024C2-\U0001F251"
        u"\U0001f926-\U0001f937"
        u"\U00010000-\U0010ffff"
        u"\u2640-\u2642" 
        u"\u2600-\u2B55"
        u"\u200d"
        u"\u23cf"
        u"\u23e9"
        u"\u231a"
        u"\ufe0f"  # dingbats
        u"\u3030"
                      "]+", re.UNICODE)
    str = re.sub(emoj, '', str)
    
    # 다시 불필요한 공백 제거
    str = str.strip(" ")
    
    # 전처리 이후에도 이상한게 남아있으면
    other = re.compile("[^\w|\'|\.|,| |’]")
    s = re.findall(other, str)
    if len(s) > 0:
        print("****error****")
    
    return str
```

```python
str = preprocess(str)
print(str)
```

```python
for s in df['Text']:
    print("------------previous------------")
    print(s)
    print("------------preprocesed------------")
    print(preprocess(s))
    print("====================================")
```
