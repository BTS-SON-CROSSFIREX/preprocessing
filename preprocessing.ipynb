```python
def preprocess(str):
    # Î∂àÌïÑÏöîÌïú Í≥µÎ∞± ÏÇ≠Ï†ú
    str = str.strip(" ")
    # ÎåÄÎ¨∏Ïûê ÏÜåÎ¨∏ÏûêÎ°ú Î∞îÍøà
    str = str.lower()
    # @Ïù¥ÎÇò RTÎ°ú ÏãúÏûëÌïòÎäî Î∂ÄÎ∂Ñ Îã§ ÏßÄÏõÄ
    while(1):
        if(str[0:1] == '@' or str[0:2] == 'RT') :
            str = ' '.join(str.split()[1:])
        else : break
    # #, @, https ÏßÄÏõÄ
    str = re.sub('#|@|http.*|\(|\)|\"|\*', '', str)
    # Í≥µÎ∞± Î¨∏Ïûê Ïã§Ï†ú Í≥µÎ∞±ÏúºÎ°ú ÎåÄÏ≤¥
    str = re.sub('\s|-|:|;|&|/', ' ', str)
    # Î¨ºÏùåÌëú, ÎäêÎÇåÌëú Ïò®Ï†êÏúºÎ°ú ÎåÄÏ≤¥
    str = re.sub('!|\?', '.', str)
    
    # Ïù¥Î™®ÏßÄ Ï†úÍ±∞
    emoj = re.compile("["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
        u"\U00002500-\U00002BEF"  # chinese char
        u"\U00002702-\U000027B0"
        u"\U00002702-\U000027B0"
        u"\U000024C2-\U0001F251"
        u"\U0001f926-\U0001f937"
        u"\U00010000-\U0010ffff"
        u"\u2640-\u2642" 
        u"\u2600-\u2B55"
        u"\u200d"
        u"\u23cf"
        u"\u23e9"
        u"\u231a"
        u"\ufe0f"  # dingbats
        u"\u3030"
                      "]+", re.UNICODE)
    str = re.sub(emoj, '', str)
    
    # Îã§Ïãú Î∂àÌïÑÏöîÌïú Í≥µÎ∞± Ï†úÍ±∞
    str = str.strip(" ")
    
    # Ï†ÑÏ≤òÎ¶¨ Ïù¥ÌõÑÏóêÎèÑ Ïù¥ÏÉÅÌïúÍ≤å ÎÇ®ÏïÑÏûàÏúºÎ©¥
    other = re.compile("[^\w|\'|\.|,| |‚Äô]")
    s = re.findall(other, str)
    if len(s) > 0:
        print("****error****")
    
    return str
```

```python
str = preprocess(str)
print(str)
```

```python
for s in df['Text']:
    print("------------previous------------")
    print(s)
    print("------------preprocesed------------")
    print(preprocess(s))
    print("====================================")
```
